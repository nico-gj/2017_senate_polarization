```python
import json
import scrapy
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import sys
from io import StringIO
```

```python
# Set Working Directory
disk = "" #C: or nothing
user = "Nico"
directory = disk+"/Users/"+user+"/Dropbox/DataPiche/Political_Clusters_Project"
```

```python
twitter_handles = pd.read_excel(directory+"/Input/Twitter_Handles.xlsx")
```

```python
nb_tweets = 20
```

```python
scraped_tweets = pd.DataFrame("", index=np.arange(nb_tweets), columns=twitter_handles['Politician'])
```

```python
for i in range(0,twitter_handles.shape[0]):
    base_url = requests.get("https://twitter.com/"+twitter_handles['Twitter_handle'][i])
    print(base_url.status_code)
    soup = BeautifulSoup(base_url.content, 'html.parser')
    tweet_feed = soup.find(id = "stream-items-id")
    tweets = tweet_feed.findAll(class_ = "TweetTextSize TweetTextSize--normal js-tweet-text tweet-text")
    temp = pd.DataFrame("", index=np.arange(nb_tweets), columns=['temp'])
    for j in range(0,nb_tweets):
        temp['temp'][j] = tweets[j].get_text()
    scraped_tweets[twitter_handles['Politician'][i]] = temp
```

```python
scraped_tweets_clean = scraped_tweets
scraped_tweets_clean = scraped_tweets_clean.replace({'—': '-'}, regex=True)
scraped_tweets_clean = scraped_tweets_clean.replace({'–': '-'}, regex=True)
scraped_tweets_clean = scraped_tweets_clean.replace({'’': '\''}, regex=True)
scraped_tweets_clean = scraped_tweets_clean.replace({'“': '\"'}, regex=True)
scraped_tweets_clean = scraped_tweets_clean.replace({'”': '\"'}, regex=True)

scraped_tweets_clean
```

```python
scraped_tweets_clean.to_csv(
    path_or_buf=directory+"/Input/Twitter_Feed_Scrape.csv"
    , sep=','
    , index = False
)
```
